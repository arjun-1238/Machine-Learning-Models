```python
import pandas as pd
df=pd.read_csv("/content/Pima indians diabetes dataset.csv")
```

```python
df.head(5)
```

```python
x=df[["BloodPressure","SkinThickness","Insulin","BMI","Age"]]
y=df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report,confusion_matrix
x_tain,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
model=GaussianNB()
model.fit(x_tain,y_train)
prediction=model.predict(x_test)
```

```python
print("Classification Report\n",classification_report(y_test,prediction))
print("Confusion Matrix\n",confusion_matrix(y_test,prediction))
```
